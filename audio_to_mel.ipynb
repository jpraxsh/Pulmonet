```python
import csv
import os
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
audpath='/content/drive/MyDrive/Respiratory_Sound_Database/audio_and_txt_files'
path='/content/drive/MyDrive/db'
patient_details={}
df=pd.read_csv('/content/drive/MyDrive/Respiratory_Sound_Database/patient_diagnosis.csv')
# print(df)
r,c=df.shape
for i in range(0,r,1):
  temp=df.iloc[i]
  # print(temp)
  patient_details[str(temp['101'])]=temp['URTI']
print(patient_details)
```

```python
def func(filename,count):
  # print(filename)
  f=filename[:3]
  # print(f)
  folder=patient_details[f]
  # print(folder)
  dir_list=os.listdir(path)
  if folder not in dir_list:
    dirname=folder
    fullpath=os.path.join(path,folder)
    os.mkdir(fullpath)
  file=audpath+'/'+filename
  y,sr = librosa.load(file)
  S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
  fig, ax = plt.subplots()
  S_dB = librosa.power_to_db(S, ref=np.max)
  img = librosa.display.specshow(S_dB, x_axis='time',y_axis='mel', sr=sr,fmax=8000, ax=ax)
  fig.colorbar(img, ax=ax, format='%+2.0f dB')
  ax.set(title='Mel-frequency spectrogram')
  plt.savefig('plot.jpg', dpi=300, bbox_inches='tight')
  plt.close()
  img = cv2.imread("plot.jpg", cv2.IMREAD_COLOR)
  height = img.shape[0]
  width = img.shape[1]
  cropped_image = img[ 93 : 1199 , 226 : 1413]
  wp = path + '/' + folder + '/'
  fwp = wp + str(count) + '.jpg'
  cv2.imwrite(fwp, cropped_image)
```

```python
arr=os.listdir(audpath)
count=1
for i in arr:
  if(i[len(i)-3:]=='wav'):
    func(i,count)
    count+=1
```
